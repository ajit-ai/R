```r
# prompt: Sample R language program

# Hello world
message("Hello world")

# variable assignment
x <- 1

# function definition
my_function <- function(y) {
  return(y + 1)
}

# call function
my_function(x)

# loops
for (i in 1:5) {
  message(i)
}

# if statement
if (x == 1) {
  message("x is 1")
} else {
  message("x is not 1")
}
```

```r

# prompt: R language for comparing two quantity or string

# define variables
quantity1 <- 10
quantity2 <- 5
string1 <- "hello"
string2 <- "world"

# comparing quantities or strings
if (quantity1 > quantity2) {
  message("quantity1 is greater than quantity2")
} else if (quantity1 < quantity2) {
  message("quantity1 is less than quantity2")
} else {
  message("quantity1 is equal to quantity2")
}

if (string1 == string2) {
  message("string1 is equal to string2")
} else {
  message("string1 is not equal to string2")
}
```

```r
#Logical Values

13 < 3
12 == 12
12 != 10
20 > 19

#using variables to compare values
x <- 20
y <- 10
x < y
x > y
x == y

```

```r
# Bar Chart

x <- c(10,20,30,40)
y <- c(30,70,80,90)
barplot(y, names.arg = x)

# change color of bar chart
barplot(y, names.arg = x, col = '#00ffff', border='red')

# change the density of bar chart
barplot(y, names.arg = x, col = '#00ffff', density = 50)

# change width of the bar chart
barplot(y, names.arg = x, width = c(3,2,1,4))

# display data horizontally
barplot(y, names.arg=x, horiz = TRUE, col='cyan', density = 75)

```

```r
#Array

#1D array
one_dim <- array(c(1:10))
one_dim

#2D array
two_dim = array(one_dim, dim=c(2, 3, 2))
two_dim

#access items
new1 = array(c(1,2,3,4),dim=c(2,2))
new1
new1[1,2]
new1[2,2]
new1[,c(1)]
new1[c(1),]

#check the existence of element/item
1 %in% new1
5 %in% new1
3 %in% new1

#number of row & col
dim(new1)
dim(two_dim)
dim(one_dim)

#length of array
length(new1)
length(one_dim)
length(two_dim)
```
